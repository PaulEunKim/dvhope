```python
pollution_df = spark.read.csv("../data/NEI_pollution_source_2020.csv", header=True, inferSchema=True)
health_df = spark.read.csv("../data/PLACES__Local_Data_for_Better_Health__County_Data_2024_release_20241023.csv", header=True, inferSchema=True)
```

```python
pollution_df.write.csv("../data_test", header=True)
```

```python
from pyspark.sql.functions import split, col, concat_ws

# Prepare Dataset 1
pollution_df = pollution_df.withColumn("StateAbbr", split(col("State-County"), " - ").getItem(0)) \
                           .withColumn("County", split(col("State-County"), " - ").getItem(1))

# Prepare Dataset 2
health_df = health_df.withColumn("County", col("LocationName"))

# Create Composite Key in Both DataFrames
pollution_df = pollution_df.withColumn("StateCountyKey", concat_ws("-", col("StateAbbr"), col("County")))
health_df = health_df.withColumn("StateCountyKey", concat_ws("-", col("StateAbbr"), col("County")))
```

```python
health_df = health_df.withColumnRenamed("County", "HealthCounty")
health_df = health_df.withColumnRenamed("stateabbr", "HealthStateabbr")
joined_df = pollution_df.join(health_df, on="StateCountyKey", how="inner")
```

```python
joined_df.printSchema()
joined_df.show(1000)
```
